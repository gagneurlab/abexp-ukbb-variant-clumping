```python
import pyranges as pr
import pandas as pd
```

```python
def get_vars_by_gene(var_mac_df, gene_id, gene_annotation, padding = 0):
    split_var_names = var_mac_df.columns[1:].to_series().str.split(":", expand=True)
    variants = pr.from_dict({
        "Chromosome":split_var_names[0].astype(str), 
        "Start":split_var_names[1].astype(int), 
        "End":split_var_names[1].astype(int)+1,
        "var_name":split_var_names.index
    })
    
    gene_annotation_expanded = gene_annoation.copy()
    gene_annotation_expanded.Start = gene_annotation_expanded.Start - padding
    gene_annotation_expanded.End = gene_annotation_expanded.End + padding
    
    included_vars = variants.intersect(gene_annotation_expanded[gene_annotation_expanded.gene_id.str.startswith(gene_id)])
    included_vars = included_vars.as_df()["var_name"].to_list() if included_vars else []
    return var_mac_df[["ID"]+included_vars]
```

```python
mac_index_vars = pd.read_parquet("/s/project/uk_biobank/processed/clumping/30780/WES_200K_index_vars_mac_30780.parquet.p")
gtf_file = "/s/genomes/Gencode/Gencode_human/release_34/gencode.v34.annotation.gtf.gz"
#gtf_file = "/s/genomes/Gencode/Gencode_human/release_34/GRCh37_mapping/gencode.v34lift37.annotation.gtf.gz"
genome_annotation = pr.read_gtf(gtf_file)
gene_annoation = genome_annotation[genome_annotation.Feature=="gene"]
```

```python
get_vars_by_gene(mac_index_vars, "ENSG00000084674", gene_annoation, padding = 150000)
```
